```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#importing data set 

fifa = pd.read_csv('C:/Users/hi/Documents/FIFA - 2022.csv')
```

```python
#lets inspect the data
#Display summary statistics and data types

fifa.info()
fifa.describe()
```

```python
#Checking for Missing Values: Identify and handle any missing data.

# Check for missing values

fifa.isnull().sum()
```

```python
#Exploring Data Characteristics: Descriptive statistics and distributions.

#Exploring Data Characteristics

import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of numerical columns

fifa.hist(bins=15, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
#Correlation Analysis: Identify relationships between variables.

#Correlation Analysis

# Correlation matrix

corr_matrix = fifa.corr()

# Heatmap of the correlation matrix

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Data Visualization: Histograms, scatter plots, and heatmaps.
#Data Visualization 

# Scatter plot of Goals For vs. Goals Against

plt.figure(figsize=(8, 6))
sns.scatterplot(data=fifa, x='Goals For', y='Goals Against', hue='Position', palette='viridis', s=100)
plt.title('Goals For vs. Goals Against')
plt.show()

```

```python

# Scatter plot of Games Played vs. Points

plt.figure(figsize=(8, 6))
sns.scatterplot(data=fifa, x='Games Played', y='Points', hue='Position', palette='viridis', s=100)
plt.title('Games Played vs. Points')
plt.show()
```

```python
#To further explore and understand the data, we can perform the following additional analyses:

#Detailed Analysis of Top Performing Teams: Investigate the characteristics of teams with high rankings.
#Outlier Detection: Identify and analyze any outliers in the dataset.
#Pairplot Visualization: Create a pairplot to observe pairwise relationships between all numerical variables.
```

```python
#1. Detailed Analysis of Top Performing Teams

# Filter top performing teams (top 5 positions)

top_teams = fifa[fifa['Position'] <= 5]
top_teams

```

```python
#2. Outlier Detection

# Boxplot to identify outliers in numerical columns

plt.figure(figsize=(15, 8))
sns.boxplot(data=fifa.drop(columns=['Position']), orient='h')
plt.title('Boxplot of Numerical Columns to Identify Outliers')
plt.show()
```

```python
#3. Pairplot Visualization

# Pairplot of numerical columns


sns.pairplot(fifa.select_dtypes(include=['int64']), diag_kind='kde')
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
plt.show()

```

```python
#Bar Chart of Points for Top Teams:

#This will show the teams with the highest points in descending order.

import matplotlib.pyplot as plt
import seaborn as sns

# Sort the dataset by Points in descending order

fifa_sorted = fifa.sort_values(by='Points', ascending=False)


```

```python
# Select columns of interest for the heatmap

columns_of_interest = ['Games Played', 'Goals For', 'Goals Against', 'Goal Difference', 'Points']
top_teams_heatmap = top_teams[columns_of_interest]

# Heatmap of performance metrics for top teams

plt.figure(figsize=(10, 6))
sns.heatmap(top_teams_heatmap, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap of Performance Metrics for Top Teams')
plt.show()

```
